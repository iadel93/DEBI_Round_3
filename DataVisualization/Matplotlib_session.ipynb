# Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y) # Line plot
plt.title('Sine Wave')
plt.xlabel('X-value')
plt.ylabel('Y-value')
# plt.grid(True)
plt.show()
```

# Bar Plot

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

plt.bar(categories, values)
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# plot it with horizontal bars
plt.barh(categories, values)
plt.title('Horizontal Bar Chart Example')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
amounts = {
    'Black': [5, 7, 3, 4],
    'White': [4, 6, 2, 5],
    'Gray': [3, 4, 5, 2]
}

width = 0.5

fig, ax = plt.subplots()
bottom = np.zeros(len(categories))

# Stacked bar plot
for color, amount in amounts.items():
    p = ax.bar(categories, amount, width, label=color, bottom=bottom)
    bottom += amount
    # break

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```

# Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

colors = ['red' if yi > 0.5 else 'blue' for yi in y]

plt.scatter(x[y>0.5], y[y>0.5], c='red', label='Y > 0.5')
plt.scatter(x[y<=0.5], y[y<=0.5], c='blue', label='Y <= 0.5')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# show legend for both colors
plt.legend(loc='upper right')
plt.show()
```

# Box Plot

```python
x  = np.random.randint(1, 100, 100)
# Adding outliers
x = np.append(x, [150, 200, 250])
print(x.mean())
plt.boxplot(x)
plt.title('Box Plot Example')
plt.ylabel('Values')
plt.show()
```

# Histogram

```python
import kagglehub
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("ammaraahmad/immigration-to-canada")

print("Path to dataset files:", path)

data = pd.read_csv(f'{path}\\canadian_immegration_data.csv')
data.info()
```

```python
imigration_count_per_year = data.melt(id_vars=['Country', 'Continent', 'Region','DevName'],
                                      var_name='Year', value_name='Count')
imigration_count_per_year = imigration_count_per_year.groupby(['Year']).sum().reset_index()
# imigration_count_per_year.head()
plt.figure(figsize=(15, 6))
plt.plot(imigration_count_per_year['Year'], imigration_count_per_year['Count'], marker='x', linestyle='--', color='purple')
plt.title('Immigration Count per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
imigration_count_per_year = data.melt(id_vars=['Country', 'Continent', 'Region','DevName'],
                                      var_name='Year', value_name='Count')
imigration_count_per_year = imigration_count_per_year.groupby(['Year', 'Continent']).sum().reset_index()
imigration_count_per_year.head()

plt.figure(figsize=(15, 6))
for continent in imigration_count_per_year['Continent'].unique():
    subset = imigration_count_per_year[imigration_count_per_year['Continent'] == continent]
    plt.plot(subset['Year'], subset['Count'], label=continent)
plt.title('Immigration Count per Year by Continent')
plt.legend(loc='upper right')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
imigration_count_per_year = data.melt(id_vars=['Country', 'Continent', 'Region','DevName'],
                                      var_name='Year', value_name='Count')
imigration_count_per_year = imigration_count_per_year.groupby(['Continent']).sum().reset_index()
imigration_count_per_year.head()

plt.figure(figsize=(10, 6))
plt.pie(imigration_count_per_year['Count'], labels=imigration_count_per_year['Continent'], autopct='%1.1f%%', startangle=140)
plt.title('Immigration Count by Continent')
plt.show()
```

```python
imigration_count_per_year = data.melt(id_vars=['Country', 'Continent', 'Region','DevName'],
                                      var_name='Year', value_name='Count')
imigration_count_per_year_continent = imigration_count_per_year.groupby(['Year', 'Continent']).sum().reset_index()
imigration_count_per_continent = imigration_count_per_year.groupby(['Continent']).sum().reset_index()

fig , ax = plt.subplots(2,1, figsize=(10,6))
for continent in imigration_count_per_year_continent['Continent'].unique():
    subset = imigration_count_per_year_continent[imigration_count_per_year_continent['Continent'] == continent]
    ax[0].plot(subset['Year'], subset['Count'], label=continent)
ax[0].set_title('Immigration Count per Year by Continent')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('Count')
ax[0].legend(loc='upper right')
ax[0].tick_params(axis='x', rotation=45)
ax[1].pie(imigration_count_per_continent['Count'], labels=imigration_count_per_continent['Continent'], autopct='%1.1f%%', startangle=140)
ax[1].set_title('Total Immigration Count per Continent')
plt.show()

```
