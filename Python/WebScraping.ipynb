```python
import requests

web_page  = requests.get("https://www.w3schools.com/sql/")
print(web_page)
print(web_page.content)
# Data retrieved is so complex
```

Beautiful Soup 

```python
import bs4
soup = bs4.BeautifulSoup(web_page.content, "html.parser")
print(soup.prettify())
```

```python
for link in soup.find_all("a"):
    print(link.get("href"))
```

```python
for text in soup.find_all("p"):
    print(text.get_text())
```

```python
# get all p elements
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode

driver = Chrome(
    options=chrome_options
)
driver.get("https://www.w3schools.com/sql/")

p_elements = driver.find_elements(By.TAG_NAME, "p")
for p in p_elements:
    print(p.get_attribute('innerText'))

driver.quit()
```

```python
# The Wuzzuf script

# get all a elements
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

chrome_options = Options()
chrome_options.add_argument("--start-maximized")
# chrome_options.add_argument("--headless")  # Run in headless mode

driver = Chrome(
    options=chrome_options
)
driver.get("https://wuzzuf.net/jobs/egypt")
driver.implicitly_wait(2)  # seconds

search_element = driver.find_element(By.XPATH, "//*[@id='app']/div/div[1]/main/div[2]/div[1]/form/div/input")
if search_element:
    search_element.send_keys("Engineer")
    search_element.send_keys(Keys.RETURN)
driver.implicitly_wait(2)  # seconds

for i in range(3):
    print(f"\n--- Page {i+1} ---")
    
    # Wait for job titles to load
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//*[@id='app']/div/div/div/div/div/div/div/div/h2/a"))
    )
    
    Job_Titles = driver.find_elements(By.XPATH, "//*[@id='app']/div/div/div/div/div/div/div/div/h2/a")
    for title in Job_Titles:
        print(title.text)

    # Try to find and click the next button
    try:
        # Wait a bit for any overlays to disappear
        time.sleep(2)
        
        # Try different selectors for the next button
        next_button = None
        
        # Try finding by class name
        next_buttons = driver.find_elements(By.CLASS_NAME, "css-wq4g8g")
        if next_buttons:
            next_button = next_buttons[-1]
        
        if next_button:
            # Scroll to the button first
            driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", next_button)
            time.sleep(1)
            
            # Try clicking with JavaScript first (more reliable)
            try:
                driver.execute_script("arguments[0].click();", next_button)
                print("Clicked next button using JavaScript")
            except Exception as js_error:
                print(f"JavaScript click failed: {js_error}")
                # Fallback to ActionChains click
                try:
                    ActionChains(driver).move_to_element(next_button).click().perform()
                    print("Clicked next button using ActionChains")
                except Exception as action_error:
                    print(f"ActionChains click failed: {action_error}")
                    # Last resort: regular click
                    try:
                        next_button.click()
                        print("Clicked next button using regular click")
                    except Exception as regular_error:
                        print(f"Regular click failed: {regular_error}")
                        break
            
            # Wait for page to load
            time.sleep(3)
        else:
            print("No next button found, might be on last page")
            break
            
    except Exception as e:
        print(f"Error on page {i+1}: {e}")
        break

print("\nScraping completed!")
# driver.quit()
```
